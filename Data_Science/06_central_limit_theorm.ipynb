```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pop_data = []
for i in range(0, 1000): 
  pop_data.append(np.random.randint(10, 100))

pop_table = pd.DataFrame({"Pop_data": pop_data})
```

```python
plt.Figure(figsize=(1,0))
sns.kdeplot(x='Pop_data', data=pop_table)
plt.show()
```

```python
sample_mean = []
for i in range(100):
  sample = []
  for data in range(500):
    sample.append(np.random.choice(pop_data))
  sample_mean.append(np.mean(sample))
```

```python
sample_mean_data = pd.DataFrame({"Sample": sample_mean})
```

```python
sns.kdeplot(x='Sample', data=sample_mean_data)
plt.show()
```
