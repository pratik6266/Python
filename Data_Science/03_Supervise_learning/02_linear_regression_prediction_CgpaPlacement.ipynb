```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('../placement.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x = dataset[['cgpa']]
y = dataset['package']
```

```python
sns.scatterplot(x='cgpa', y='package', data=dataset)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict([[6.89]])
```

```python
lr.score(x_test, y_test)
```

```python
m = lr.coef_
m[0]
```

```python
c = lr.intercept_
c
```

```python
y = m[0] * 6.8 + c # This is the y = mx + c equation according to line eqution and this model has been created
y 
```

```python
sns.scatterplot(x='cgpa', y='package', data=dataset, label='original')
plt.plot(dataset['cgpa'], lr.predict(dataset[['cgpa']]),c='red', label='predicted')
plt.legend()
plt.show()
```
