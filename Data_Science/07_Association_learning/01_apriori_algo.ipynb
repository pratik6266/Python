```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import collections
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
dataset = pd.read_csv('../groceries.csv')
dataset.head(3)
```

```python
dataset.shape
```

```python
market = []
for i in range(0, dataset.shape[0]):
  cus = []
  for j in dataset.columns:
    if type(dataset[j][i]) == str:
      cus.append(dataset[j][i])
  market.append(cus)
```

```python
collections.Counter([])
```

```python
l = []
for i in market:
  for j in i:
    l.append(j)
```

```python
p = collections.Counter(l)
```

```python
d = {"Items name": p.keys(), "val": p.values()}
d
```

```python
pf = pd.DataFrame(d).sort_values(by='val', ascending=False)
```

```python
pf
```

```python
tr = TransactionEncoder()
tr.fit(market)
new_df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
```

```python
new_df
```

```python
apriori(new_df, min_support=0.05, use_colnames=True).sort_values(by=['support'])
```
