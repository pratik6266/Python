```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import collections
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
from mlxtend.frequent_patterns.fpgrowth import fpgrowth
```

```python
dataset = pd.read_csv('../groceries.csv')
```

```python
dataset.shape
```

```python
market = []
for i in range(0, dataset.shape[0]):
  cus = []
  for j in dataset.columns:
    if type(dataset[j][i]) == str:
      cus.append(dataset[j][i])
  market.append(cus)
```

```python
l = []
for i in market:
  for j in i:
    l.append(j)
```

```python
l
```

```python
p = collections.Counter(l)
```

```python
d = {"Items": p.keys(), "Val": p.values()}
d
```

```python
pf = pd.DataFrame(d).sort_values(by=['Val'], ascending=False)
pf
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
new_df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
```

```python
new_df.shape
```

```python
fpgrowth(new_df, min_support=0.07, use_colnames=True, max_len=3).sort_values(by=['support']) 
```
