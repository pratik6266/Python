```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('../placement2.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x='cgpa', y='resume_score', data=dataset, hue='placed')
plt.axvline(x=6.5, c='red')
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset['placed']
```

```python
y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
plt.Figure(figsize=(2,1))
sns.kdeplot(data=dataset['cgpa'], label='cgpa', linestyle='dashed')
sns.kdeplot(data=dataset['resume_score'], label='resume_score', c='red', linestyle=':')
plt.legend()
plt.show()
```

```python
gnb = GaussianNB() # uses when data follow gaussian curve
gnb.fit(x_train, y_train)
gnb.score(x_test, y_test) * 100
```

```python
bnb = BernoulliNB() # uses when data has boolean data only like one hot encoded where there are 2 data
bnb.fit(x_train, y_train)
bnb.score(x_test, y_test) * 100
```

```python
mnb = MultinomialNB() # uses when multiclass data are there like categorical data
mnb.fit(x_train, y_train)
mnb.score(x_test, y_test) * 100
```

```python
plot_decision_regions(X=x.to_numpy(), y=y.to_numpy(), clf=gnb)
plt.show()
```

```python
plot_decision_regions(X=x.to_numpy(), y=y.to_numpy(), clf=bnb)
plt.show()
```

```python
plot_decision_regions(X=x.to_numpy(), y=y.to_numpy(), clf=mnb)
plt.show()
```
