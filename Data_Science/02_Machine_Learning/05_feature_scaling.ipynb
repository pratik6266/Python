```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# we do feature scaling becaue we want the less magnitude feature to be there as well so scalling bring down the upper and lower so that they come close but they do so in respective manner that's how nature don't changes.
dataset = pd.read_csv('../loan.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
ss = StandardScaler()
ss.fit(dataset[['ApplicantIncome']])
```

```python
ar = ss.transform(dataset[['ApplicantIncome']])
dataset['ApplicantIncome_ss'] = pd.DataFrame(ar, columns=['x'])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.distplot(dataset['ApplicantIncome_ss'])
plt.plot()
```

```python
dataset.isnull().sum()
```

```python
mm = MinMaxScaler()
mm.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_mm'] = mm.transform(dataset[['CoapplicantIncome']])
```
