```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.svm import SVC
```

```python
x, y = make_moons(n_samples=1000, noise=0.2)
```

```python
df = {
  "data1": x[:, 0],
  "data2": x[:, 1]
}
dataset = pd.DataFrame(df)
```

```python
dataset.head(2)
```

```python
sns.scatterplot(x='data1', y='data2', data=dataset, hue=y)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state=42)
```

```python
bg = BaggingClassifier(estimator=SVC(), n_estimators=30)
bg.fit(x_train, y_train)
```

```python
bg.score(x_test, y_test)*100, bg.score(x_train, y_train)*100
```

```python
rf = RandomForestClassifier(n_estimators=30)
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test)*100, rf.score(x_train, y_train)*100
```
