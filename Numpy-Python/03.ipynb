```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
sum_col = np.sum(arr, axis=0)
print(sum_col)
new = np.sum(sum_col, axis=0)
print(new)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
cumsum = np.cumsum(arr, axis=0)
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(cumsum)
plt.title("Average")
plt.xlabel("years")
plt.ylabel("sales")
plt.show()
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([6, 7, 8, 9])

print(arr1 + arr2)
print(arr1 * arr2)
print(np.dot(arr1, arr2))
```
