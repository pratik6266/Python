```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['Mango', 'Banana', 'Apple', 'Orange', 'Melon']
y = [10, 20, 24, 22, 35]
crack = [0, 0, 0, 1, 0]
cmap = plt.get_cmap('inferno')
colors = cmap(np.linspace(0, 1, len(y)))
```

```python
plt.pie(y, labels=x, explode=crack, autopct='%1.1f%%', colors=colors)
plt.show()
```
